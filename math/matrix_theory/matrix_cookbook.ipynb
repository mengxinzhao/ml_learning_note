## 第 1 章：基础（Basics）

基本恒等式:

$$(AB)^{-1} = B^{-1}A^{-1} \tag{1}$$

$$(ABC...)^{-1} = ...C^{-1}B^{-1}A^{-1} \tag{2}$$

$$(A^T)^{-1} = (A^{-1})^T \tag{3}$$

$$(A+B)^T = A^T + B^T \tag{4}$$

$$(AB)^T = B^T A^T \tag{5}$$

$$(ABC...)^T = ...C^T B^T A^T \tag{6}$$

$$(A^H)^{-1} = (A^{-1})^H \tag{7}$$

$$(A+B)^H = A^H + B^H \tag{8}$$

$$(AB)^H = B^H A^H \tag{9}$$

$$(ABC...)^H = ...C^H B^H A^H \tag{10}$$

### 1.1 迹（Trace）

> **几何意义**：迹是矩阵所有特征值之和，即线性变换在各特征方向上拉伸量的总和。它衡量的是变换对空间的"总体缩放趋势"——正迹表示整体在膨胀，负迹表示整体在收缩。
>
> 在微分方程 $\dot{x} = Ax$ 中，$\text{Tr}(A)$ 直接决定了相空间体积的变化率（Liouville 定理）：$\frac{d}{dt}\text{Vol} = \text{Tr}(A) \cdot \text{Vol}$。在机器学习中，Hessian 矩阵的迹（即 Laplacian）常用于衡量损失函数曲面的平均曲率。

$$\text{Tr}(A) = \sum_i A_{ii} \tag{11}$$

$$\text{Tr}(A) = \sum_i \lambda_i, \lambda_i = \text{eig}(A) \tag{12}$$

$$\text{Tr}(A) = \text{Tr}(A^T) \tag{13}$$

$$\text{Tr}(AB) = \text{Tr}(BA) \tag{14}$$

$$\text{Tr}(A+B) = \text{Tr}(A) + \text{Tr}(B) \tag{15}$$

$$\text{Tr}(ABC) = \text{Tr}(BCA) = \text{Tr}(CAB) \tag{16}$$

$$\mathbf{a}^T\mathbf{a} = \text{Tr}(\mathbf{a}\mathbf{a}^T) \tag{17}$$

> **Note**
>
> **(16) 迹的循环不变性（Cyclic Invariance）：** 只能循环移位，不能任意换序。$\text{Tr}(ABC) = \text{Tr}(CAB)$，但一般 $\text{Tr}(ABC) \neq \text{Tr}(ACB)$。
>
> **(17) 内积 = 外积的迹：** $\mathbf{a}^T\mathbf{a}$ 是内积（标量），$\mathbf{a}\mathbf{a}^T$ 是外积（矩阵），外积的对角元素为 $a_i^2$，所以 $\text{Tr}(\mathbf{a}\mathbf{a}^T) = \sum_i a_i^2 = \mathbf{a}^T\mathbf{a}$。
>
> $$\mathbf{a}\mathbf{a}^T = \begin{bmatrix} a_1^2 & a_1 a_2 & \cdots \\ a_2 a_1 & a_2^2 & \cdots \\ \vdots & \vdots & \ddots \end{bmatrix} \implies \text{Tr}(\mathbf{a}\mathbf{a}^T) = a_1^2 + a_2^2 + \cdots = \mathbf{a}^T\mathbf{a}$$
>
> 在机器学习中非常常用，例如推导岭回归、正则化损失的梯度时，经常需要把 $\|\mathbf{w}\|^2 = \mathbf{w}^T\mathbf{w}$ 写成 $\text{Tr}(\mathbf{w}\mathbf{w}^T)$，因为迹对矩阵求导有很方便的公式（见第 2.5 节）。


### 1.2 行列式（Determinant）

> **几何意义**：矩阵 $A$ 把单位超立方体（unit hypercube）映射成一个平行多面体（parallelepiped），$|\det(A)|$ 就是这个平行多面体的体积，而 $\det(A)$ 的符号表示变换是否翻转了空间的定向（orientation）。

设 $A$ 为 $n \times n$ 矩阵。

$$\det(A) = \prod_i \lambda_i, \quad \lambda_i = \text{eig}(A) \tag{18}$$

$$\det(cA) = c^n \det(A), \text{ if } A \in \mathbb{R}^{n\times n} \tag{19}$$

$$\det(A^T) = \det(A) \tag{20}$$

$$\det(AB) = \det(A)\det(B) \tag{21}$$

$$\det(A^{-1}) = 1/\det(A) \tag{22}$$

$$\det(A^n) = \det(A)^n \tag{23}$$

$$\det(I + \mathbf{u}\mathbf{v}^T) = 1 + \mathbf{u}^T\mathbf{v} \tag{24}$$

当 $n=2$：

$$\det(I+A) = 1 + \det(A) + \text{Tr}(A) \tag{25}$$

当 $n=3$：

$$\det(I+A) = 1 + \det(A) + \text{Tr}(A) + \frac{1}{2}\text{Tr}(A)^2 - \frac{1}{2}\text{Tr}(A^2) \tag{26}$$

当 $n=4$：

$$\det(I+A) = 1 + \det(A) + \text{Tr}(A) + \frac{1}{2}\text{Tr}(A)^2 - \frac{1}{2}\text{Tr}(A^2) + \frac{1}{6}\text{Tr}(A)^3 - \frac{1}{2}\text{Tr}(A)\text{Tr}(A^2) + \frac{1}{3}\text{Tr}(A^3) \tag{27}$$

当 $\varepsilon$ 很小时：

$$\det(I + \varepsilon A) \approx 1 + \det(A) + \varepsilon\text{Tr}(A) + \frac{1}{2}\varepsilon^2\text{Tr}(A)^2 - \frac{1}{2}\varepsilon^2\text{Tr}(A^2) \tag{28}$$

> **记忆方法（公式 25–28）**
>
> 核心是：$\det(I+A) = \prod_i (1+\lambda_i)$，然后把括号乘开。
>
> 以 $n=3$，三个特征值 $\lambda_1, \lambda_2, \lambda_3$ 为例：
>
> $$\det(I+A) = (1+\lambda_1)(1+\lambda_2)(1+\lambda_3)$$
>
> 暴力展开：
>
> $$= 1 + \underbrace{(\lambda_1+\lambda_2+\lambda_3)}_{\text{选1个}} + \underbrace{(\lambda_1\lambda_2+\lambda_1\lambda_3+\lambda_2\lambda_3)}_{\text{选2个}} + \underbrace{\lambda_1\lambda_2\lambda_3}_{\text{全选}}$$
>
> 首尾两项已经认识：
>
> | 展开项 | 等于 |
> |-------|---------|
> | $\lambda_1+\lambda_2+\lambda_3$ | $\text{Tr}(A)$ |
> | $\lambda_1\lambda_2\lambda_3$ | $\det(A)$ |
> | $\lambda_1\lambda_2+\lambda_1\lambda_3+\lambda_2\lambda_3$ | 需要技巧 $\downarrow$ |
>
> 中间项用"平方差"技巧：
>
> $$(\lambda_1+\lambda_2+\lambda_3)^2 = \lambda_1^2+\lambda_2^2+\lambda_3^2 + 2(\lambda_1\lambda_2+\lambda_1\lambda_3+\lambda_2\lambda_3)$$
>
> $$\Rightarrow \lambda_1\lambda_2+\lambda_1\lambda_3+\lambda_2\lambda_3 = \frac{1}{2}\left[\left(\sum \lambda_i\right)^2 - \sum \lambda_i^2\right] = \frac{1}{2}\left[\text{Tr}(A)^2 - \text{Tr}(A^2)\right]$$
>
> 代回去就得到公式 (26)。$n=2$ 时没有中间项所以最简单，$n$ 越大中间项越多，但推导方式一样。$\varepsilon$ 近似只保留到 $\varepsilon^2$ 阶（即最多"选2个"）。

### 1.3 特例：$2\times 2$ 矩阵

考虑矩阵 $A = \begin{bmatrix} A_{11} & A_{12} \\ A_{21} & A_{22} \end{bmatrix}$

**行列式与迹：**

$$\det(A) = A_{11}A_{22} - A_{12}A_{21} \tag{29}$$

$$\text{Tr}(A) = A_{11} + A_{22} \tag{30}$$

**特征值：**

$\lambda^2 - \lambda \cdot \text{Tr}(A) + \det(A) = 0$

$\lambda_1 = \frac{\text{Tr}(A) + \sqrt{\text{Tr}(A)^2 - 4\det(A)}}{2}, \quad \lambda_2 = \frac{\text{Tr}(A) - \sqrt{\text{Tr}(A)^2 - 4\det(A)}}{2}$

$\lambda_1 + \lambda_2 = \text{Tr}(A), \quad \lambda_1\lambda_2 = \det(A)$

**特征向量：**

$\mathbf{v}_1 \propto \begin{bmatrix} A_{12} \\ \lambda_1 - A_{11} \end{bmatrix}, \quad \mathbf{v}_2 \propto \begin{bmatrix} A_{12} \\ \lambda_2 - A_{11} \end{bmatrix}$

**逆矩阵：**

$$A^{-1} = \frac{1}{\det(A)}\begin{bmatrix} A_{22} & -A_{12} \\ -A_{21} & A_{11} \end{bmatrix} \tag{31}$$


```python
import numpy as np
np.random.seed(42)

A = np.random.randn(3, 3)
B = np.random.randn(3, 3)
C = np.random.randn(3, 3)

print("Cyclic invariance: Tr(ABC) = Tr(BCA) = Tr(CAB)  ... (16)")
print(f"  Tr(ABC) = {np.trace(A @ B @ C):.6f}")
print(f"  Tr(BCA) = {np.trace(B @ C @ A):.6f}")
print(f"  Tr(CAB) = {np.trace(C @ A @ B):.6f}")

print(f"\nNote: Tr(ABC) != Tr(ACB) in general (cyclic shift only, not arbitrary reorder)")
print(f"  Tr(ACB) = {np.trace(A @ C @ B):.6f}")

print(f"\nInner product = Trace of outer product: a^T a = Tr(aa^T)  ... (17)")
a = np.random.randn(4, 1)
print(f"  a^T a    = {(a.T @ a)[0,0]:.6f}")
print(f"  Tr(aa^T) = {np.trace(a @ a.T):.6f}")
```

## 第 2 章：导数（Derivatives）

本节讨论对矩阵 $X$ 的表达式求导。始终假定 $X$ 无特殊结构（各元素相互独立）。关于结构化矩阵，见第 2.8 节。

**基本假设：**

$$\frac{\partial X_{kl}}{\partial X_{ij}} = \delta_{ik}\delta_{lj} \tag{32}$$

**一般求导法则：**

$A$ 为常数时：

$$\partial A = 0 \tag{33}$$

$$\partial(\alpha X) = \alpha \partial X \tag{34}$$

$$\partial(X+Y) = \partial X + \partial Y \tag{35}$$

$$\partial(\text{Tr}(X)) = \text{Tr}(\partial X) \tag{36}$$

$$\partial(XY) = (\partial X)Y + X(\partial Y) \tag{37}$$

$$\partial(X \circ Y) = (\partial X) \circ Y + X \circ (\partial Y) \tag{38}$$

$$\partial(X \otimes Y) = (\partial X) \otimes Y + X \otimes (\partial Y) \tag{39}$$

$$\partial(X^{-1}) = -X^{-1}(\partial X)X^{-1} \tag{40}$$

$$\partial(\det(X)) = \text{Tr}(\text{adj}(X)\partial X) \tag{41}$$

$$\partial(\det(X)) = \det(X)\text{Tr}(X^{-1}\partial X) \tag{42}$$

$$\partial(\ln(\det(X))) = \text{Tr}(X^{-1}\partial X) \tag{43}$$

$$\partial X^T = (\partial X)^T \tag{44}$$

$$\partial X^H = (\partial X)^H \tag{45}$$

**辨认法则（Identification）：** 对于标量函数 $f(X)$，微分与导数的关系为：

$$df = \text{Tr}\!\left(\left(\frac{\partial f}{\partial X}\right)^T dX\right)$$

类比标量微积分中 $df = f'(x)\,dx$：标量情况下从 $df = 3\,dx$ 直接读出 $f'=3$；矩阵情况下，将 $df$ 凑成 $\text{Tr}(\cdots\, dX)$ 的形式，$dX$ 前面的系数就是 $\left(\frac{\partial f}{\partial X}\right)^T$。

为什么可以直接"读出"系数？因为 $\langle A, B \rangle_F = \text{Tr}(A^T B) = \sum_{ij} A_{ij} B_{ij}$ 是矩阵空间上的 Frobenius 内积（就是把矩阵对应元素相乘再求和，跟向量点积完全类比）。上面的公式用内积写就是 $df = \langle \frac{\partial f}{\partial X},\, dX \rangle_F$。在内积空间中，若对所有 $dX$ 都有 $\langle G, dX \rangle = \langle H, dX \rangle$，则 $G = H$，所以系数唯一确定。

> **Note**
>
> **(40) 逆矩阵的微分：** 类比一元函数 $d(y^{-1}) = -y^{-2} \cdot dy$，矩阵版本是 $\partial(X^{-1}) = -X^{-1}(\partial X)X^{-1}$。标量情形中 $y^{-2} = y^{-1} \cdot y^{-1}$，矩阵情形中也是"左乘 $X^{-1}$，右乘 $X^{-1}$"。
>
> 推导：从恒等式 $X X^{-1} = I$ 两边取微分，由乘法法则 (37)：
>
> $$(\partial X)X^{-1} + X\,\partial(X^{-1}) = 0$$
>
> 解出 $\partial(X^{-1})$：
>
> $$\partial(X^{-1}) = -X^{-1}(\partial X)X^{-1}$$
>
>
> **(41–43) 行列式的微分：**
>
> 伴随矩阵（adjugate）的定义：$\text{adj}(X)$ 满足 $X \cdot \text{adj}(X) = \det(X) \cdot I$，当 $X$ 可逆时 $\text{adj}(X) = \det(X) \, X^{-1}$。
>
> **(41)** 的推导：余子式（cofactor）$\text{cof}(X,i,j) = (-1)^{i+j}\det([X]_{ij})$，其中 $[X]_{ij}$ 是删掉第 $i$ 行第 $j$ 列后的 $(n{-}1)\times(n{-}1)$ 子矩阵（详见 3.1.2 节 (146)）。行列式按第一行展开为 $\det(X) = \sum_j X_{1j} \, \text{cof}(X,1,j)$。由于 $\text{cof}(X,i,j)$ 不含 $X_{ij}$（那行那列已被删除），所以 $\frac{\partial \det(X)}{\partial X_{ij}} = \text{cof}(X,i,j)$。将所有偏导组装成矩阵就是 $\text{cof}(X)^T = \text{adj}(X)$。用迹的形式写出：
>
> $$\partial(\det(X)) = \sum_{ij} \frac{\partial \det(X)}{\partial X_{ij}} \partial X_{ij} = \sum_{ij} \text{adj}(X)_{ji} \, \partial X_{ij} = \text{Tr}(\text{adj}(X) \, \partial X)$$
>
> **(42)** 当 $X$ 可逆时，代入 $\text{adj}(X) = \det(X) X^{-1}$：
>
> $$\partial(\det(X)) = \text{Tr}(\det(X) X^{-1} \partial X) = \det(X) \, \text{Tr}(X^{-1} \partial X)$$
>
> **(43)** 由链式法则 $\partial(\ln f) = \frac{\partial f}{f}$，将 (42) 两边除以 $\det(X)$：
>
> $$\partial(\ln(\det(X))) = \frac{\partial(\det(X))}{\det(X)} = \text{Tr}(X^{-1} \partial X)$$
>

## 第 3 章：逆矩阵（Inverses）

### 3.1 基础

#### 3.1.1 定义

矩阵 $A \in \mathbb{C}^{n \times n}$ 的逆 $A^{-1}$ 定义为满足下式的矩阵。若 $A^{-1}$ 存在，则称 $A$ 为非奇异矩阵；否则为奇异矩阵。

$$AA^{-1} = A^{-1}A = I \tag{145}$$

#### 3.1.2 余子式与伴随矩阵

子矩阵 $[A]_{ij}$ 为删去第 $i$ 行与第 $j$ 列后得到的 $(n-1) \times (n-1)$ 矩阵。

余子式：

$$\text{cof}(A,i,j) = (-1)^{i+j}\det([A]_{ij}) \tag{146}$$

伴随矩阵：

$$\text{adj}(A) = (\text{cof}(A))^T \tag{148}$$

#### 3.1.3 行列式

$$\det(A) = \sum_{j=1}^n (-1)^{j+1}A_{1j}\det([A]_{1j}) \tag{149}$$

#### 3.1.4 构造

$$A^{-1} = \frac{1}{\det(A)}\text{adj}(A) \tag{151}$$

#### 3.1.5 条件数

$$c(A) = \frac{d_+}{d_-} \tag{152}$$

$$c(A) = \|A\| \cdot \|A^{-1}\| \tag{153}$$

### 3.2 精确关系式

#### 3.2.1 基本

$$(AB)^{-1} = B^{-1}A^{-1} \tag{155}$$

#### 3.2.2 Woodbury 恒等式

$$(A+CBC^T)^{-1} = A^{-1} - A^{-1}C(B^{-1}+C^TA^{-1}C)^{-1}C^TA^{-1} \tag{156}$$

$$(A+UBV)^{-1} = A^{-1} - A^{-1}U(B^{-1}+VA^{-1}U)^{-1}VA^{-1} \tag{157}$$

若 $P$、$R$ 正定，则

$$(P^{-1}+B^TR^{-1}B)^{-1}B^TR^{-1} = PB^T(BPB^T+R)^{-1} \tag{158}$$

#### 3.2.3 Kailath 变体

$$(A+BC)^{-1} = A^{-1} - A^{-1}B(I+CA^{-1}B)^{-1}CA^{-1} \tag{159}$$

#### 3.2.4 Sherman–Morrison

$$(A+\mathbf{b}\mathbf{c}^T)^{-1} = A^{-1} - \frac{A^{-1}\mathbf{b}\mathbf{c}^TA^{-1}}{1+\mathbf{c}^TA^{-1}\mathbf{b}} \tag{160}$$

#### 3.2.5 Searle 恒等式组

$$(I+A^{-1})^{-1} = A(A+I)^{-1} \tag{161}$$

$$A - A(A+B)^{-1}A = B - B(A+B)^{-1}B \tag{164}$$

$$(I+AB)^{-1}A = A(I+BA)^{-1} \tag{167}$$
